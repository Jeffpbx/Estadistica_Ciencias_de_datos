```python
article.parse() #Dividir el texto por cada palabra 

autor = ", ".join(author for author in article.authors)
titulo = article.title
fecha = article.publish_date
texto = article.text
imagenes = article.top_image
videos = article.movies
url = article.url

print("Informacion del articulo")
print("=" * 30)
print(f"Titulo: {titulo}")
print(f"Autor(s): {autor}")
print(f"Fecha de Publicacion: {fecha}")
print(f"Imagen: {imagenes}")
print(f"Videos: {videos}")
print(f"Link del articulo: {url}")
print(f"Contenido: {texto[:100] + '...'}")
```
